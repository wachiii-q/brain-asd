```python
# Import necessary libraries
import os
import numpy as np
import mne

# Define the input and output directories
input_dir = "/Users/wachiii/Workschii/brain-asd/temp/processed_teee"
output_dir = "/Users/wachiii/Workschii/brain-asd/temp/processed_epochs"
os.makedirs(output_dir, exist_ok=True)  # Create the output directory if it doesn't exist

# Define constants
DATASET_SFREQ = 256  # Sampling frequency (Hz)
EPOCH_LENGTH = 30  # Epoch length in seconds

# Loop through all .npy files in the input directory
for file_name in os.listdir(input_dir):
    if file_name.endswith(".npy"):  # Process only .npy files
        file_path = os.path.join(input_dir, file_name)
        print(f"Processing file: {file_path}")
        
        # Load the .npy file
        eeg_data = np.load(file_path)

        # Define channel names and types
        ch_names = [f"Ch{i+1}" for i in range(eeg_data.shape[0])]
        ch_types = ["eeg"] * eeg_data.shape[0]

        # Create MNE Raw object
        info = mne.create_info(ch_names=ch_names, sfreq=DATASET_SFREQ, ch_types=ch_types)
        raw = mne.io.RawArray(eeg_data, info)

        # Apply preprocessing
        raw.filter(l_freq=0.5, h_freq=40, fir_design='firwin')
        raw.notch_filter(freqs=50, fir_design='firwin')

        # Define events for the entire 30-second file
        events = np.array([[0, 0, 1]])  # Single event at the start of the file
        event_id = {"Dummy": 1}

        # Create epochs
        tmin = 0
        tmax = EPOCH_LENGTH - 1 / DATASET_SFREQ  # Adjust for sampling frequency
        epochs = mne.Epochs(raw, events, event_id, tmin=tmin, tmax=tmax, baseline=None, preload=True)

        # Save the epochs as a .fif file
        output_file = os.path.join(output_dir, file_name.replace(".npy", ".fif"))
        epochs.save(output_file, overwrite=True)
        print(f"Saved epochs to: {output_file}")
```

```python
# Import necessary libraries
import os
import mne

# Define the input and output directories
input_dir = "/Users/wachiii/Workschii/brain-asd/temp/processed_epochs"
output_dir = "/Users/wachiii/Workschii/brain-asd/temp/merged_epochs"
os.makedirs(output_dir, exist_ok=True)  # Create the output directory if it doesn't exist

# Group files by their base names (e.g., H08_EC, H08_EO)
file_groups = {}
for file_name in os.listdir(input_dir):
    if file_name.endswith(".fif"):  # Process only .fif files
        base_name = "_".join(file_name.split("_")[:2])  # Extract base name (e.g., H08_EC)
        if base_name not in file_groups:
            file_groups[base_name] = []
        file_groups[base_name].append(file_name)

# Merge files in each group
for base_name, files in file_groups.items():
    print(f"Merging files for: {base_name}")
    epochs_list = []
    for file_name in sorted(files):  # Sort to ensure consistent order
        file_path = os.path.join(input_dir, file_name)
        print(f"Loading file: {file_path}")
        epochs = mne.read_epochs(file_path, preload=True)
        epochs_list.append(epochs)

    # Concatenate epochs
    merged_epochs = mne.concatenate_epochs(epochs_list)

    # Save the merged epochs
    output_file = os.path.join(output_dir, f"{base_name}.fif")
    merged_epochs.save(output_file, overwrite=True)
    print(f"Saved merged file to: {output_file}")
```
