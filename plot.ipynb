```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracy data (in %)
band_power_acc = [70.56, 90.56, 70.00, 89.44, 87.78]
relative_power_acc = [69.44, 82.22, 80.56, 88.33, 75.00]

# Model names
models = ['Decision Tree', 'Random Forest', 'Logistic Regression', 'Gradient Boosting', 'AdaBoost']

# Custom colors for each model
colors = {
    'Decision Tree': 'lightsteelblue',
    'Random Forest': 'darksalmon',
    'Logistic Regression': 'darkseagreen',
    'Gradient Boosting': 'lightblue',
    'AdaBoost': 'plum'
}

# Custom x positions: extra spacing between Band and Relative groups
x_band = np.array([0, 1, 2, 3, 4])/1.5
x_relative = x_band + 4

# Create the plot
plt.figure(figsize=(14, 7))
bar_width = 0.5

# Plot each model bar in Band and Relative groups
for i, model in enumerate(models):
    plt.bar(x_band[i], band_power_acc[i], width=bar_width, color=colors[model])
    plt.bar(x_relative[i], relative_power_acc[i], width=bar_width, color=colors[model])

# Set ticks and labels
x_ticks = np.concatenate((x_band, x_relative))
x_labels = [f'{model}\n(Band)' for model in models] + [f'{model}\n(Relative)' for model in models]
plt.xticks(ticks=x_ticks, labels=x_labels, rotation=45, ha='right')

# Labels and title
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracy Comparison: Band Power vs Relative Power', pad=20)

# Custom legend
handles = [plt.Rectangle((0, 0), 1, 1, color=colors[model]) for model in models]
plt.legend(handles, models, title='Model', loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=3)

# Grid and spacing
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.subplots_adjust(left=0.07, right=0.98, top=0.9, bottom=0.25)

plt.show()
```
